````python
a=numpy.array([1,2,3])
print(a.shape)
b=numpy.array([[1,2,3]])
print(b.shape)
print(a,b)

c=numpy.zeros((3,3))
print(c)
print(numpy.concatenate((b,c),axis=0))
````
